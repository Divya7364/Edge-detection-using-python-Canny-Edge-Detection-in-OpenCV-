# Canny Edge Detection in OpenCV
* This algorithm present in the Open-cv library
* The canny edge detector is an edge detection operator that uses  multi-stage algorithm to detect a wide range of edges in images

```python
img=cv2.imread("/content/drive/MyDrive/medium-doremon-.webp",0)
edges=cv2.Canny(img,100,200)
```

```python
plt.subplot(121),plt.imshow(img,cmap='gray')
plt.title('Original image'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap='gray')
plt.title('Edge image'),plt.xticks([]),plt.yticks([])
plt.show()
```
